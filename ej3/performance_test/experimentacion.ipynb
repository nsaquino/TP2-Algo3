# Experimentación del TP2-Ej3

```python
#Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

from pathlib import Path
output_path = os.path.join(Path.cwd(), 'output')
```

### Comparación entre instancias

```python
df = pd.read_csv(output_path + "/runtime_tp_noOpt.csv")
df_onlyPath = pd.read_csv(output_path + "/runtime_tp_onlyPath.csv")
df_rank_noPath = pd.read_csv(output_path + "/runtime_tp_rank_noPath.csv")
df_rank_Path = pd.read_csv(output_path + "/runtime_tp_rank_path.csv")
df_size_noPath = pd.read_csv(output_path + "/runtime_tp_size_noPath.csv")
df_size_path = pd.read_csv(output_path + "/runtime_tp_size_path.csv")

df['time_noOpt'] = df['time']
df['time_onlyPath'] = df_onlyPath['time']
df['time_rank_noPath'] = df_rank_noPath['time']
df['time_rank_Path'] = df_rank_Path['time']
df['time_size_noPath'] = df_size_noPath['time']
df['time_size_path'] = df_size_path['time']

fig, ax = plt.subplots()

ax.scatter(df['n'],df['time_noOpt'],color='yellow')
ax.plot(df['n'],df['time_noOpt'],color='yellow', label='noOpt')

ax.scatter(df['n'],df['time_onlyPath'],color='blue')
ax.plot(df['n'],df['time_onlyPath'],color='blue', label='onlyPath')

ax.scatter(df['n'],df['time_rank_noPath'],color='red')
ax.plot(df['n'],df['time_rank_noPath'],color='red', label='rank_noPath')

ax.scatter(df['n'],df['time_rank_Path'],color='red')
ax.plot(df['n'],df['time_rank_Path'],color='red', label='rank_Path')

ax.scatter(df['n'],df['time_size_noPath'],color='green')
ax.plot(df['n'],df['time_size_noPath'],color='green', label='size_noPath')

ax.scatter(df['n'],df['time_size_path'],color='black')
ax.plot(df['n'],df['time_size_path'],color='black', label='size_path')

plt.xlabel("n")
plt.ylabel("tiempo (s)")
ax.legend()
ax.grid(True)
```
